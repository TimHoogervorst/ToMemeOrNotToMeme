```python
from PIL import Image
import os

def get_image_dimensions(image_path):
    try:
        with Image.open(image_path) as img:
            width, height = img.size
            return width * height
    except Exception as e:
        print(f"Error reading image {image_path}: {e}")
        return None

def check_folder_for_image_dimensions(folder_path):
    image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        dimensions = get_image_dimensions(image_path)

        if dimensions:
            print(f"Image: {image_file}, Dimensions: {dimensions}")

if __name__ == "__main__":
    folder_path = "./train/memes/"
    
    if os.path.exists(folder_path):
        check_folder_for_image_dimensions(folder_path)
    else:
        print(f"The specified folder does not exist: {folder_path}")
```
